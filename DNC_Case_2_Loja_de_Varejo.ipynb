### TRATAMENTO DE DADOS

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

- Como APP e Aplicativo é o mesmo, será feita a substituição de um dos dois para unificar o dado.

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique() #ok
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

- Substituir os nomes com espaço por underline:

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

- Tratamento dos nulos (estados)
  - Eliminar linhas inteiras?
  - Substituir dado?

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

- Porém neste caso não funciona por conta da premissa que nos estados com nulo deveria se aplicar 'MS' assim:

```python
#vendas['estado'] = vendas['estado'].fillna('MS')
vendas['estado'].fillna('MS', inplace=True) #faz a mesma coisa que a sintaxe de cima
vendas.info()
```

- Quanto aos preços, ao invés de se deletar as linhas pode ser interessante utilizar a média dos valores para preencher.

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace=True)
```

```python
vendas.info()
```

### FILTROS

```python
 vendas[['Preço','Preço_com_frete']].describe()
```

- Trabalhando a premissa de que o valor do Preço não pode ser maior do que o valor do Preço_com_frete
- Assim serão eliminadas as linhas em que acontece isso
- Duas opções:
  - Alterar a base de dados para que selecione somente as linhas onde não acontece isso
  - Eliminar as linhar onde acontece

```python
preco_errado = vendas.query('Preço > Preço_com_frete')

vendas_correto = vendas.query('Preço < Preço_com_frete') #ideal seria <=, mas vou manter conforme o vídeo
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

- Agora por exemplo (Filtro treinar) que se queira ver analisar os itens de Esporte e Lazer no estado de SP:

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

### Análise das Métricas

```python
#Departamentos mais vendidos
  #Agrupa por Nome_Departamento, com o critério id.compra
  #Quantidade nunique()
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#Média de preço com frete por departamento
  #Duas casas decimais função round
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
#Quantidade de vendas por mês
vendas_correto.groupby('Data').idcompra.nunique().reset_index()
```

```python
#Para aparecer por mês efetivamente:
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().reset_index()
```

## Unindo Diferentes Bases
- Para responder às outras duas métricas é necessário importar o a outra base e realizar o join das tabelas

### Join das tabelas

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
```

```python
cliente.head()
```

- A conexão pode ser realizada pela coluna cliente_log

```python
cliente.info()
cliente[['idade', 'renda']].describe()
```

```python
#Transformando renda para float
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
#Unindo tabelas
vendas_cliente = vendas_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
#Média de renda para cada tipo de canal de venda
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
#Média de idade de clientes por bandeira
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index())
agg_idade_bandeira
```

## Transformando em visualizações

```python
#Importando bibliotecas
import matplotlib.pyplot as plt
```

```python
#Idade média por Bandeira
#Tamanhos do gráfico
fig = plt.figure(figsize=(7,4))

#Gráfico de barra
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')

#Mudando o label da idade
plt.ylabel('Média da idade')

#Colocando um título, com formatações para ele
plt.title('Idade média por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
#Média de renda para cada tipo de canal de venda
#Tamanhos do gráfico
fig = plt.figure(figsize=(7,4))

#Gráfico de barra
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='green')

#Mudando o label da categoria
plt.xlabel('Categoria')
plt.ylabel('Renda')

#Colocando um título, com formatações para ele
plt.title('Média de renda por Canal de Venda', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
#Mudança com o tempo
venda_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data['Data']
```

```python
fig = plt.figure(figsize=(20,12))

plt.xlabel('Data da venda')
plt.ylabel('Quantidade de vendas')
x = venda_por_data['Data']
y = venda_por_data['idcompra']

plt.plot(x, y, color='green', linewidth=2)
plt.show()
```

```python
#Média de preço com frete por departamento

agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
#Tamanhos do gráfico
fig = plt.figure(figsize=(20,8))

#Gráfico de barra
plt.bar(agg_dept_preco['Nome_Departamento'], agg_dept_preco['Preço_com_frete'], color='green')

#Mudando o label da categoria
plt.xlabel('Nome do Departamento')
plt.ylabel('Preço com Frete')

#Colocando um título, com formatações para ele
plt.title('Média de preço com frete por departamento', loc='center', fontsize=18, fontweight='bold', color='black');
```
